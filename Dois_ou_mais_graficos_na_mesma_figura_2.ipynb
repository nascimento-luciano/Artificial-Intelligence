```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({'font.size': 14})

x = np.linspace(-1.55,1.55,1000)

plt.subplot(2,2,1)
plt.plot(x,np.sin(x),'--k',linewidth=2)
plt.subplot(2,2,2)
plt.plot(x,np.cos(x),':b',linewidth=2)
plt.subplot(2,1,2)
plt.plot(x,np.tan(x),'-.r',linewidth=2)
plt.tight_layout()
```

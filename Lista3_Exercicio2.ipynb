b) Compare o resultado com os labels conhecidos destas flores. Qual foi a acuraria do k-means?

```python
#Importando bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
from sklearn.metrics.cluster import adjusted_rand_score

#Iris
iris=load_iris()

#Plotar Iris
X=iris.data[:, :4]
Y=iris.target
print(np.shape(X))
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.scatter(X[:,0], X[:,1], c=Y)
plt.xlabel('sepal length'); plt.ylabel('sepal Width')
plt.title('data set, cores são os clusters verdadeiros')
plt.subplot(1,2,2)
plt.scatter(X[:,2], X[:,3],c=Y)
plt.xlabel('petal length');plt.ylabel('petal Width')
plt.title('data set, cores são os clusters verdadeiros')

#Junta a entradas em clusters
kmeans=KMeans(n_clusters=3, max_iter=200, random_state=1).fit(X)

#Acha os centros de cada clusters
centros=kmeans.cluster_centers_

#Plota o gráfico dos clusters em 3D
fig=plt.figure()
ax=Axes3D(fig)
ax.scatter(X[:,0], X[:,1], X[:,2], c=kmeans.labels_,alpha=0.4)                   #Transparência dos dados
ax.plot(centros[:,0], centros[:,1], centros[:,2], 'rx', markersize=8, label='centros dos clusters')
ax.set_xlabel('x1');ax.set_ylabel('x2');ax.set_zlabel('x3')
plt.legend()

print(Y)
print(kmeans.labels_)

acuracia=adjusted_rand_score(Y,kmeans.labels_)
print('rand score:', acuracia)







```

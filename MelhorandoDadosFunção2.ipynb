```python
import numpy as np                                    #importa a biblioteca numpy com o apelido de np
import matplotlib.pyplot as plt                       #importa a biblioteca matplotlib com o apelido de plt

plt.rcParams.update({'font.size':14})                 #fixa em 14 todas a fonte para tudo no gráfico

x = np.linspace(-3,5,1000)                            #cria o vetor [-3, -2, -1, 0, 1, 2, 3, 4, 5] e guarda na variável "x"
y = x*x-2*x+4                                         #calcula x.x-2.x+4 e guarda na variável "y"        
z = x**3-3*x+10                                       #calcula x..x3.3.x+10 e guarda na variável "z"   

plt.plot(x,y,'--k',label='parábola',linewidth=2)      #o texto 'o--k': coloca uma marca redonda, linha tracejada e preta  
plt.plot(x,z,'-r',label='função cúbica',linewidth=2)  #o texto 'r': coloca uma marca redonda, linha vermelha  
plt.title('Duas Curvas')                              #coloca o título no gráfico, "Gráfico bonito"
plt.xlabel('x')                                       #coloca legenda no eixo x, chamada "x"
plt.ylabel('função')                                  #coloca uma legenda no eixo y, os $ fazem o texto ser interpretado como equação!
plt.xlim([-5, 6])                                     #define que o gráfico irá exibir x entre -5 e 6
plt.ylim([0, 20])                                     #define que o gráfico irá exibir y entre 0 e 20
plt.legend(loc=4)
plt.grid()                                            #coloca grade no gráfico
```

```python
import numpy as np                                    #importa a biblioteca numpy com o apelido de np
import matplotlib.pyplot as plt                       #importa a biblioteca matplotlib com o apelido de plt

plt.rcParams.update({'font.size':14})                 #fixa em 14 todas a fonte para tudo no gráfico

x = np.linspace(-3,5,1000)                            #cria o vetor [-3, 5, 1000] e guarda na variável "x"
y = x*x-2*x+4                                         #calcula x.x-2.x+4 e guarda na variável "y"        
z = x**3-3*x+10                                       #usamos uma nova variável para a segunda função.lembre: em python x**3 é x ao cubo

plt.plot(x,y,'--k',label='parábola',linewidth=2)      #o label ='parábola' será usado na legenda
plt.plot(x,z,'-r',label='função cúbica',linewidth=2)  #a segunda curva será vermelha linha sólida 
plt.title('Duas Curvas')                              #coloca o título no gráfico, "Duas curvas"
plt.xlabel('x')                                       #coloca legenda no eixo x, chamada "x"
plt.ylabel('função')                                  #coloca uma legenda no eixo y, chamada "y"
plt.xlim([-5, 6])                                     #define que o gráfico irá exibir x entre -5 e 6
plt.ylim([0, 20])                                     #define que o gráfico irá exibir y entre 0 e 20
plt.legend(bbox_to_anchor=(1.04,1))                   #bbox_to_anchor=(1.04,1) define a localização da legenda para fora da figura
#plt.legend(loc=4)                                    #o loc =4 define a localização da legenda
plt.grid()                                            #coloca grade no gráfico
```

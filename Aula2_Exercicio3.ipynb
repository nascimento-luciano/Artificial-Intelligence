```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 10})
x = np.linspace(-3,5,1000)
p1 = 2*x**4-4*x**2+2
p2 = -3*x**3+5*x
p3 = 2*x**2-4*x**2+2
p4 = -3*x+5
plt.subplot(1,2,1)
plt.plot(x,p1,'--k',label='$2*x^4-4*x^2+2$',linewidth=2)
plt.plot(x,p2,':r',label='$-3*x^3+5*x$',linewidth=2)
plt.plot(x,p3,'-b',label='$2*x^2-4*x^2+2$',linewidth=2)
plt.plot(x,p4,'-.m',label='$-3*x+5$',linewidth=2)
plt.legend()
plt.xlabel('x')
plt.ylabel('polin√¥mios')
plt.ylim([-30,100])
plt.legend(loc=1)

plt.grid()
plt.subplot(4,2,2)
plt.plot(x,p1,'--k',label='$2*x^4-4*x^2+2$',linewidth=2)
plt.legend()
plt.subplot(4,2,4)
plt.plot(x,p2,':r',label='$-3*x^3+5*x$',linewidth=2)
plt.legend()
plt.subplot(4,2,6)
plt.plot(x,p3,'-b',label='$2*x^2-4*x^2+2$',linewidth=2)
plt.legend()
plt.subplot(4,2,8)
plt.plot(x,p4,'-.m',label='$-3*x+5$',linewidth=2)
plt.legend()
plt.tight_layout()
```

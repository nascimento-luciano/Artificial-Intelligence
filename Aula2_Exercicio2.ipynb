```python
import numpy as np                                      #importa a biblioteca numpy com o apelido de np
import matplotlib.pyplot as plt                         #importa a biblioteca matplotlib com o apelido de plt

plt.rcParams.update({'font.size':14})                   #fixa em 14 todas a fonte para tudo no gráfico

x = np.linspace(-1.55,1.55,100)                         #cria o vetor [-3, 5, 1000] e guarda na variável "x"

plt.plot(x,x,'-k',label='linear',linewidth=2)           #plota o gráfico da função linear
plt.plot(x,np.sin(x),':b',label='seno',linewidth=2)     #plota o gráfico da função seno
plt.plot(x,np.cos(x),'--r',label='cosseno',linewidth=2) #plota o gráfico da função cosseno
plt.legend(loc=4)                                       #plota a legenda: linear, seno e cosseno
plt.xlabel('x')


```

```python
import numpy as np                                    #importa a biblioteca numpy com o apelido de np
import matplotlib.pyplot as plt                       #importa a biblioteca matplotlib com o apelido de plt

plt.rcParams.update({'font.size':14})                 #fixa em 14 todas a fonte para tudo no gráfico

x = np.linspace(-3,5,5000)                            #cria o vetor [-3, 5, 1000] e guarda na variável "x"
y = x*x-2*x+4                                         #calcula x.x-2.x+4 e guarda na variável "y"        
z = x**3-3*x+10                                       #usamos uma nova variável para a segunda função.lembre: em python x**3 é x ao cubo
plt.subplot(1,2,1)
plt.plot(x,y,'--k',linewidth=2)                       #o label ='parábola' será usado na legenda
plt.title('parábola')                                 #coloca o título no gráfico, "parábola"
plt.xlabel('x')                                       #coloca legenda no eixo x, chamada "x"
plt.subplot(1,2,2)                                       #define que o gráfico irá exibir x entre -5 e 6
plt.plot(x,z,'-r',linewidth=2)                        #o label ='parábola' será usado na legenda
plt.title('função cúbica')                            #coloca o título no gráfico, "função cúbica"
plt.xlabel('x')                                       #coloca legenda no eixo x, chamada "x"
plt.tight_layout()
```

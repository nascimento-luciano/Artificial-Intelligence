```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({'font.size': 14})

x = np.linspace(-1,1,1000)

plt.subplot(2,2,1)
plt.plot(x,x,'--k',linewidth=2)
plt.subplot(2,2,2)
plt.plot(x,x**2,':b',linewidth=2)
plt.subplot(2,2,3)
plt.plot(x,x**3,'-.r',linewidth=2)
plt.subplot(2,2,4)
plt.plot(x,x**4,'-m',linewidth=2)
plt.tight_layout()
```
